## Visualization

**Q.1. Which City has the Highest restaurants and which has the Least?**


```python
City=data.City.value_counts()
plt.bar(City.index,City.values,color='blue')
plt.title("Restaurant counts by city",fontsize=3,fontweight='bold')
plt.xticks(rotation=90)
plt.xlabel("Cities")
plt.ylabel("Counts")
plt.show()

```

## Insights: Kolkata has the **Highest** number of restaurants that co-works with Swiggy and Surat has the **Least**

**Q.2. Display the frequency of average rating appropriate chart.**

```python
# Create a Histogram

plt.hist(data['Avg ratings'])
plt.xlabel('Average Ratings')
plt.ylabel('Frequency os ratings')
plt.title('Frequency of Average Ratings',fontsize=3,fontweight='bold')
plt.show()
```

**Q.3. What is the top 10 ordered food type in Swiggy?**

```python
# Create a bar chart

food=data['Food type'].value_counts().head(10)
plt.bar(food.index,food.values)
plt.xticks(rotation=90)
plt.title('Top 10 Ordered Food types',fontsize=3,fontweight='bold')
plt.xlabel("Foodtypes")
plt.ylabel("Order Count")
plt.show()
```

##Insight: **Indian**foodtype has the most number of orders in swiggy and **beverages** has the least.

**Q.4. Which are the top 10 restaurants with fast delivery in Bangalore city?**

```python
# Create a horizontal barchart

a=data.City=='Bangalore'
b=data[a]
c=b.sort_values('Delivery time').head(10)
plt.figure(figsize=(8,5))
plt.barh(c.Restaurant[::-1],c['Delivery time'][::-1])
plt.title("Top 10 restaurants with fastest delivery in Bangalore",fontsize=3,fontweight='bold')
plt.xlabel('Time in minutes')
plt.ylabel('Restaurant Names')
plt.show()
```

##Insight: **MTR** and **MCdonalds** are the restaurants with fastest delivery time in Bangalore City.
