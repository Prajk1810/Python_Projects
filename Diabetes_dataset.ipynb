```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/diabetes-dataset.csv")
```

```python
data.head()
```

```python
data.replace({"Up":1,"Down":2,"Steady":0,"No":3},inplace=True)
```

```python
data["diabetes"].replace({"Yes":1,3:0},inplace=True)
```

```python
data.replace({"max_glu_serum":{">200":2, ">300": 3, "None": 0, "Norm": 1},
            "A1Cresult": {">7": 1, ">8": 2, "None": 0, "Norm": 3},
            "change": {3: 0, "Ch": 1},
            "readmitted": {">30":1, "NO": 0}},
           inplace = True)
```

```python
data.head()
```

```python
data.drop(columns=["patient_nbr"], inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
## Bulding a model##

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
##Spiltting the data##

X=data.drop(columns=['diabetes'])
Y=data['diabetes']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
rf_classifier=RandomForestClassifier(random_state=42)
rf_classifier
```

```python
rf_classifier.fit(X_train,Y_train)
```

```python
Y_pred=rf_classifier.predict(X_test)
```

```python
accuracy=accuracy_score(Y_test,Y_pred)
accuracy
```

```python
print(classification_report(Y_test,Y_pred))
```

```python
# Confusion Matrix#

cm=confusion_matrix(Y_test,Y_pred)
sns.heatmap(cm,annot=True,fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
```
